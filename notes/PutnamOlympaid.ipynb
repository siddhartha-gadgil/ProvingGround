<h1>Proof Generated By Lemma Propagation</h1>
<h2><u>Problem Statement</u></h2>
<p>Let $*$ be a commutative and associative binary operation on a set $S.$ Assume that for every $x$ and $y$ in $S,$ there exists $z$ in $S$ such that $x*z=y.$ (This $z$ may depend on $x$ and $y.$) Show that if $p,q,r$ are in $S$ and $p*r=q*r,$ then $p=q.$
</p>

<h2><u>Definitions</u></h2>
<ol>
    <li><b>Identity Element</b>- A element $e$ is a <b>Identity Element</b> if $\exists x \in S$ such that $x*e=x$.</li>
    <li><b>Universal Identity</b>-A <b>Identity Element</b> $e$ is called a <b>Universal Identity</b> if $\forall x\in S \ \   x*e=x $   is satisfied.</li>
</ol>

<h2><u>Solving Strategy</u></h2>
<h3>Part1 : Proving the existence and uniqueness of universal identity</h3>
<ol>
    <li>Prove that all identity elements are universal identities for the set $ S $ $i.e$ </li> $\\$
             $$\exists   x \  \exists  e \  [(x*e  =  e)  \implies \forall  y  \in S \ (y*e \ = \ y)]\\$$
<li>Prove that the set of all universal identities is singleton $i.e$ </li> $\\$
$$\forall x \ \exists e_{1} \exists e_{2} \ [[(x*e_{1} = x) \land (x*e_{2} =x )] \implies (e_{1} = e_{2})]$$ 

</ol>
<h3>Part2 : Elimination</h3>
<p>We have now proved the exists a unique <b>Universal Identity</b>, namely $e$ for the set $S$.Now for every pair $(x,e)$ we can find a $y$ (depending on $x$ and $e$) such that $x*y=e$ .Using this observation we can eliminate $c$ from the equation $a*c=b*c$.
</p>

<p><b>In the section below we define the following operators</b>:-
<ol>
    <li><b>eqM</b> - This operation is defined on set $S$ and is equivalent to the Equality operator $i.e$, eqM$($a$)($b$) \iff a=b$.Bascillay treat it as truth assignment.</li>
    <li><b>op</b> -  This operation is defined on set $S$ and is equivalent to the $*$ operation $i.e$,  $op($a$)($b$) \iff  a*b$.</li>
    <li><b>op_2</b> - This opeartion is defined on the set $S$ and is used to encompass the associative nature of the  binary operation $*$.op_2($a$)($b$)($c$) returns a term $a*b*c$.Now since the operation $*$ is associative therfore any order of bracketting will evalute the same result $i.e$, $(a*b)*c = a*(b*c)$.This is done to avoid the extra information of bracketing.</li>
    <li><b>inv</b> - This opeartion is defined on the set $S$ and takes in elements say $x$ and $y$ as inputs and returns the element $z$ such that $x*z=y$ is satisfied.We know that such a mapping exists because of the condition given in the question.</li>
</ol>
</p>

```scala
val eqM = "eqM" :: M ->: M ->: Type
val op  = "op" :: M ->: M ->: M
val op_2 = "op_2" :: M ->: M ->: M ->: M
val inv = "inv" :: M ->: M ->: M
```

<h4>Step7:</h4>
$$(y*e) = y$$

Basically we wil prove
$$[ \{ (x*e)*inv(x)(y) = (y*e) \} \land \{x*inv(x)(y) = y\} \land \{(x*e)*inv(x)(y)) = ((x*inv(x)(y))\} ] \implies (y*e=y)$$
$$\Updownarrow$$
$$\{(a=b) \land (c=d) \land (a=c)\} \implies (b=d)$$
where 
$$ a= (x*e)*inv(x)(y) \\
   b= (y*e)  \\
   c= x*inv(x)(y) \\
